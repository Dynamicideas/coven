```python
#the model that we that we will be using is mbart

from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
```

```python
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-one-to-many-mmt")


tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-one-to-many-mmt", src_lang="en_XX")


#we will be using mbart tokenizer instead of our regular autotokenizer we use before.
```

```python
word_en = "The world is a beautiful place, make friends and not enemies"
```

```python
model_inputs = tokenizer(word_en, return_tensors="pt")
```

```python
model_inputs

#we can see that our inputs have been converted to tokens
```

```python
#lets translate to french.
# where we use the language code for french to be "fr_XX", you can simply use another language code. just check the mbart code for languages supported yet.

generated_tokens = model.generate(
    **model_inputs,
    forced_bos_token_id=tokenizer.lang_code_to_id["fr_XX"]
)
```

```python
translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```
