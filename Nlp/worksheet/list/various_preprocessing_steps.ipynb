```python
#we will be using the nltk package for data cleaning an preprocessing.
#we will first covert our data to lower case.

df_eng['clean_text'] = df_eng['text'].str.lower()

df_eng.head()

df_yor['clean_texty'] = df_yor[2].str.lower()

df_yor.head()
```

```python
#here will will try to remove the punctuations marks we have in our data
#lets create a function that we can call to do that.

def remove_punctuations(text):
    punctuations = string.punctuation
    return text.translate(str.maketrans('', '', punctuations))
```

```python
df_eng['clean_text']= df_eng['clean_text'].apply(lambda x: remove_punctuations(x))
df_eng.head()
```

```python
df_yor['clean_texty']= df_yor['clean_texty'].apply(lambda x: remove_punctuations(x))
df_yor.head()
```

```python
from typing_extensions import Text
#lets create a function we can call again.

STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(Text):
    return " ".join([word for word in Text.split() if word not in STOPWORDS])

#assuming there is a yoruba language selection in nltk package. our function will look like this below
    STOPWORD = set(stopwords.words('yoruba'))
def remove_stopword(text):
    return " ".join([word for word in text.split() if word not in STOPWORD])
```

```python
df_eng['clean_text'] = df_eng['clean_text'].apply(lambda x: remove_stopwords(x))
df_eng.head()
```

```python
#next, we can also remove frequent words.


from collections import Counter
word_count = Counter()
for clean_text in df_eng['clean_text']:
    for word in clean_text.split():
        word_count[word] += 1

word_count.most_common(10)
```

```python
#lets create a function for frequent words removal.

FREQUENT_WORDS = set(word for (word, wc) in word_count.most_common(10))
def remove_freq_words(text):
    return " ".join([word for word in text.split() if word not in FREQUENT_WORDS])
```

```python
df_eng['clean_text'] = df_eng['clean_text'].apply(lambda x: remove_freq_words(x))
df_eng.head()
```

```python
#lets see if frequent words can be removed in the yoruba language too.

from collections import Counter
word_count = Counter()
for text in df_yor['clean_texty']:
    for word in text.split():
        word_count[word] += 1

word_count.most_common(20)


```

```python

FREQUENT_WORD = set(word for (word, wc) in word_count.most_common(20))
def remove_freq_word(text):
    return " ".join([word for word in text.split() if word not in FREQUENT_WORD])
```

```python
df_yor['clean_texty'] = df_yor['clean_texty'].apply(lambda x: remove_freq_word(x))
df_yor.head()


#we can see frequent word removal function works in the Yoruba language too.
```

```python
#lets try and remove words that are not common

UNCOMMON = set(word for (word, wc) in word_count.most_common()[:-10:-1])
UNCOMMON
```

```python
def remove_uncommon(text):
    return " ".join([word for word in text.split() if word not in UNCOMMON])
```

```python
df_eng['clean_text'] = df_eng['clean_text'].apply(lambda x: remove_uncommon(x))
df_eng.head()


#we can use the same code for yoruba corpus too
```

```python
#Using Stemming as another method of preprocessing

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
df_eng['stem_text'] = df_eng['clean_text'].apply(lambda x: stem_words(x))
df_eng.head()

#we can see that what stemming does is to cut bit part out of the words.. observe the stem_text column below.
#stemming is being used depending on the type of datasets that we are dealing or working on.
```

```python
#lets checkout lemmatization
#alos with POS tagging..POS means part of speech tagging. using noun, verb.etc to tag words

from nltk import pos_tag
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
wordnet_map = {"N": wordnet.NOUN, "V": wordnet.VERB, "A": wordnet.ADJ, "R": wordnet.ADV}



```

```python

    wordnet.ADJ
```

```python
def lemmatize_words(text):

    pos_text = pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_text])


```

```python
df_eng['lemmatized_text'] = df_eng['clean_text'].apply(lambda x: lemmatize_words(x))
df_eng.head()

#if we check at the lemmatized column. we will see that some words has been repaired or make meaniful and better.
```
